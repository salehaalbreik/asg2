```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
# Group 16 data
# Vector Data
a = np.array([16, 8])
b = np.array([16, 18])

# Matrix data
A = np.array([[2, 5], [1, 6]])
B = np.array([[1, 2], [0, 4]])

# Transformation data
V = np.array([10, 15])
theta = 40
S = np.array([[8, 0, 0], [0, 3, 0], [0, 0, 1]])
T = np.array([[1, 0, 2], [0, 1, 3], [0, 0, 1]])
```

```python
# Vectors a and b
def draw_vec(ax, vec, color, label):
    ax.arrow(0,0, vec[0], vec[1], head_width=0.6, length_includes_head=True, color=color)
    ax.text(vec[0]*1.02, vec[1]*1.02, label, color=color)

fig1, ax1 = plt.subplots(figsize=(6,6))
ax1.set_aspect('equal')
ax1.axhline(0,color='k'); ax1.axvline(0,color='k')
draw_vec(ax1, a, 'b', 'a')
draw_vec(ax1, b, 'r', 'b')
ax1.set_xlim(-2,35); ax1.set_ylim(-2,30)
ax1.set_title("Vectors a and b")
plt.show()
```

```python
#	Find difference between the vectors
diff = a-b
print(diff)
```

```python
# Find the sum of vectors
sum = a+b
print(sum)
```

```python
# 	Find the angles α and θ using trig function
def vector_angle(v):
    return np.degrees(np.arctan2(v[1], v[0]))

alpha = vector_angle(a)
theta_vec = vector_angle(b)
print(f"α  = {alpha:.2f}°, θ = {theta_vec:.2f}°")
```

```python
#	Euclidean norms ‖a‖,‖b‖
norm_a = np.linalg.norm(a)
norm_b = np.linalg.norm(b)
print(f"||a|| = {norm_a:.3f}, ||b|| = {norm_b:.3f}")
```

```python
# 6. Dot product
dot_product = np.dot(a, b)
print(f"a • b = {dot_product}")
```

```python
# Dot product using magnitudes and angle
angle_between = np.arccos(dot_product / (norm_a * norm_b))
dot_product_alt = norm_a * norm_b * np.cos(angle_between)
print(f"a • b =  |a||b|cosθ = {dot_product_alt:.2f}")
print(f"Angle between vectors: {np.degrees(angle_between):.2f}°")
```

```python

#Draw the parallelogram which is described by the two vectors

O = np.array([0,0])
AB = a+b
poly_x = [O[0], a[0], AB[0], b[0], O[0]]
poly_y = [O[1], a[1], AB[1], b[1], O[1]]

fig3, ax3 = plt.subplots(figsize=(6,6))
ax3.set_aspect('equal')
ax3.axhline(0,color='k'); ax3.axvline(0,color='k')
draw_vec(ax3, a, 'b', 'a')
draw_vec(ax3, b, 'r', 'b')
ax3.plot(poly_x, poly_y, 'k--')
ax3.set_xlim(-2,35); ax3.set_ylim(-2,30)
ax3.set_title("Parallelogram spanned by a and b")
plt.show()
```

```python

# Cross product
a3 = np.array([16, 8, 0])
b3 = np.array([16, 18, 0])
cross_product = np.cross(a3, b3)
print(f"a × b = {cross_product}")
```

```python

# Cross product
cross_magnitude = norm_a * norm_b * np.sin(angle_between)
print(f"axb =  |a||b|sinθ = {cross_magnitude:.2f}")
```

```python
from mpl_toolkits.mplot3d import Axes3D

# Create 3D figure
fig = plt.figure(figsize=(6,6))
ax = fig.add_subplot(111, projection='3d')

# Draw vector a (blue)
ax.quiver(0,0,0, a3[0], a3[1], a3[2], color='b')
ax.text(a3[0], a3[1], a3[2], "a", color='b')

# Draw vector b (red)
ax.quiver(0,0,0, b3[0], b3[1], b3[2], color='r')
ax.text(b3[0], b3[1], b3[2], "b", color='r')

# Draw cross product (green, along z)
ax.quiver(0,0,0, cross_product[0], cross_product[1], cross_product[2], color='g')
ax.text(cross_product[0], cross_product[1], cross_product[2], "a×b", color='g')

# Axes limits
ax.set_xlim(0,20)
ax.set_ylim(0,20)
ax.set_zlim(0,180)

# Labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")
ax.set_title("Cross product a×b (Group 16)")

plt.show()
```

```python
#Transformations
# 1. Scaled vector
V_homogeneous = np.append(V, 1)
V_scaled_homogeneous = S @ V_homogeneous
V_scaled = V_scaled_homogeneous[:2]
print(f"V' = S × V = {V_scaled}")


plt.figure(figsize=(6,6))
plt.axhline(0,color='k'); plt.axvline(0,color='k')
plt.arrow(0,0, V_scaled[0], V_scaled[1], head_width=1.0, color='g', length_includes_head=True)
plt.text(V_scaled[0]*1.02, V_scaled[1]*1.02, f"v_s({V_scaled[0]:.1f},{V_scaled[1]:.1f})", color='g')
plt.xlim(-10,100); plt.ylim(-10,100)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Scaled vector v_s = S v")
plt.xlabel("X"); plt.ylabel("Y")
plt.show()
```

```python
#  Rotated vector
vh = np.append(V, 1)
theta_rad = np.deg2rad(theta)
R = np.array([[np.cos(theta_rad), -np.sin(theta_rad), 0],
              [np.sin(theta_rad),  np.cos(theta_rad), 0],
              [0,0,1]])
v_r = R @ vh
v_r2d = v_r[:2]

print(f"2. Rotated vector V' = R × V ≈ ({v_r2d[0]:.2f}, {v_r2d[1]:.2f})")

plt.figure(figsize=(6,6))
plt.axhline(0,color='k'); plt.axvline(0,color='k')
plt.arrow(0,0, v_r2d[0], v_r2d[1], head_width=1.0, color='r', length_includes_head=True)
plt.text(v_r2d[0]*1.05, v_r2d[1]*1.05, f"v_r≈({v_r2d[0]:.1f},{v_r2d[1]:.1f})", color='r')
plt.xlim(-10,100); plt.ylim(-10,100)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Rotated vector v_r = R v (θ=40°)")
plt.xlabel("X"); plt.ylabel("Y")
plt.show()
```

```python
#V' = (RS)V

v_rs = R @ S @ vh
v_rs2d = v_rs[:2]

plt.figure(figsize=(6,6))
plt.axhline(0,color='k'); plt.axvline(0,color='k')
plt.arrow(0,0, v_rs2d[0], v_rs2d[1], head_width=1.0, color='m', length_includes_head=True)
plt.text(v_rs2d[0]*1.02, v_rs2d[1]*1.02, f"v_rs≈({v_rs2d[0]:.1f},{v_rs2d[1]:.1f})", color='m')
plt.xlim(-10,100); plt.ylim(-10,100)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Scaled + Rotated vector v_rs = R S v")
plt.xlabel("X"); plt.ylabel("Y")
plt.show()
```

```python
# Compute H = T R S
T = np.array([[1,0,2],
              [0,1,3],
              [0,0,1]])

H = T @ R @ S
print("Homogeneous Transformation Matrix H =\n", H)


```

```python
# V'=HV
v_prime = H @ vh
print("Transformed vector v' = H v =\n", v_prime)
```

```python
v_p2d = v_prime[:2]

plt.figure(figsize=(6,6))
plt.axhline(0,color='k'); plt.axvline(0,color='k')

# Plot original vector for reference
plt.arrow(0,0, V[0], V[1], head_width=1.0, color='b', length_includes_head=True)
plt.text(V[0]*1.05, V[1]*1.05, "Original v", color='b')

# Plot final transformed vector
plt.arrow(0,0, v_p2d[0], v_p2d[1], head_width=1.0, color='g', length_includes_head=True)
plt.text(v_p2d[0]*1.02, v_p2d[1]*1.02, f"v'≈({v_p2d[0]:.1f},{v_p2d[1]:.1f})", color='g')

plt.xlim(-10,100); plt.ylim(-10,100)
plt.gca().set_aspect('equal', adjustable='box')
plt.title("Final transformed vector v' = (T R S) v")
plt.xlabel("X"); plt.ylabel("Y")
plt.show()
```
